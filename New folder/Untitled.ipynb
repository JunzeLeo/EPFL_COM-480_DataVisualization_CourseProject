```python
import pandas as pd
```

```python
data = pd.read_csv("new.csv")
```

```python
newData = pd.DataFrame(columns=data.columns)
for c in data.country.unique():
    country = data[data["country"] == c]
    
    urbans = country.urban
    urbans = (urbans-min(urbans)) / (max(urbans) - min(urbans))
    country['urban'] = urbans
    
    temps = country.temp
    temps = (temps-min(temps))/(max(temps)-min(temps))
    country['temp'] = temps
    newData = newData.append(country)
```

```python
newData.sort_values(by=['year','country'], inplace=True)
newData
```

```python
newData.to_csv("newData.csv")
```
